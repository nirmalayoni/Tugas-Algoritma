##**Persamaan Kuadrat**

```python
# persamaan garis scr umum => ax^2 + bx + c = 0
# dengan a tdk sama dengan 0
# penyelesaian persamaan kuadrat ada 3 (rumus abc,pemfaktoran,melengkapkan kuadrat sempurna)
# rumus abc => (-b +- sqrt D)/2a
# hasil kali akar akar x1x2 = c/a
# jumlah hasil akar x1+x2 = -b/a
# 1. ax^2 + bx + c = 0
# 2. x^2 - (x1+x2)x + x1x2 = 0
```

```python
# Kasus 1 = Menentukan persamaan kuadrat jika diketahui akar-akarnya (x1 dan x2)
import cmath

intro_07 = print("Konten Persamaan Kuadrat^-^: Menentukan persamaan kuadrat jika diketahui akar-akarnya (x1 dan x2)")

# inisialisasi nilai akar-akar persamaan kuadrat x1 dan x2
x1_07 = -7
x2_07 = 3

# menampilkan nilai akar-akar
print('Nilai akar x1_07 = ', x1_07)
print('Nilai akar x2_07 = ', x2_07)

#y = x*x - (x1_07+x2_07)x + x1_07*x2_07
# membentuk persamaan kuadrat
a_07 = 1  # sebagai koefisien dari x^2
b_07 = -(x1_07 + x2_07)  # sebagai koefisien dari x
c_07 = x1_07 * x2_07  # sebagai konstanta

# Menampilkan persamaan kuadrat
if b_07 > 0 and c_07 > 0:
    print("Persamaan kuadrat yang dibentuk y = x^2 + {}x + {}".format(b_07,c_07))
elif b_07 < 0 and c_07 > 0:
    print("Persamaan kuadrat yang dibentuk y = x^2 {}x + {}".format(b_07,c_07))
elif b_07 > 0 and c_07 < 0:
    print("Persamaan kuadrat yang dibentuk y = x^2 + {}x {}".format(b_07,c_07))
else:
    print("Persamaan kuadrat yang dibentuk y = x^2 {}x {}".format(b_07,c_07))
```
