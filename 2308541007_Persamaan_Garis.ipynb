## **1. Persamaan Garis Sederhana**

```python
# persamaan garis sederhana 1
import numpy as np
import matplotlib.pyplot as plt
# misalkan diketahui persamaan garis yaitu y=2x-1
# diberikan titik-titik x dan y tanpa diketahui gradiennya
x_07 = np.array([0,3,6,9])
y_07 = np.array([-1,5,11,17])

# menunjukkan titik tertentu pada sumbu koordinat
xticks_07 = plt.xticks([0,3,6,9])
yticks_07 = plt.yticks([-1,5,11,17])
# memberi label untuk x dan y
xlabel_07 = plt.xlabel('Sumbu X')
ylabel_07 = plt.ylabel('Sumbu Y')
# menambahkan garis kotak ke dalam plot
grid_07 = plt.grid()
# membuat garis lurus dengan memplot setiap titik x dan y
garis_07 = plt.plot(x_07,y_07,marker='.',ms='10',mfc='black',mec='black')
# memberi judul bentuk persamaan garis yang terbentuk
title_07 = plt.title('Persamaan Garis: y=2x-1')
# menampilkan persamaan garis yang terbentuk
tampil_07 = plt.show()
```
