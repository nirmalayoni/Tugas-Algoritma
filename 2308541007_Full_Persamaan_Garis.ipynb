## **Persamaan Garis**

```python
import numpy as np
import matplotlib.pyplot as plt

judul1_07 = print("Konten Dua Persamaan Garis")

# persamaan garis pertama yang melewati titik (a,b) dan (c,d)
# titik (a,b) sebagai (x1,y1)
a_07 = int(input("Coba masukkan titik x1 garis pertama= "))
b_07 = int(input("Coba masukkan titik y1 garis pertama= "))
# titik (c,d) sebagai (x2,y2)
c_07 = int(input("Coba masukkan titik x2 garis pertama= "))
d_07 = int(input("Coba masukkan titik y2 garis pertama= "))

# persamaan garis kedua yang melewati titik (e,f) dan (g,h)
# titik (e,f) sebagai (x1,y1)
e_07 = int(input("Coba masukkan titik x1 garis kedua= "))
f_07 = int(input("Coba masukkan titik y1 garis kedua= "))
# titik (g,h) sebagai (x2,y2)
g_07 = int(input("Coba masukkan titik x2 garis kedua= "))
h_07 = int(input("Coba masukkan titik y2 garis kedua= "))

# mencari masing-masing gradien
m1_07 = (d_07-b_07)/(c_07-a_07) # gradien garis pertama
m2_07 = (h_07-f_07)/(g_07-e_07) # gradien garis kedua

# Persamaan garis pertama adalah y1 = m1(x-a)+b = m1x-m1a+b
# Persamaan garis kedua adalah y2 = m2(x-e)+f = m2x-m2e+f

c1_07 = -m1_07*a_07+b_07 # Titik potong garis pertama terhadap sumbu y
c2_07 = -m2_07*e_07+f_07 # titik potong garis kedua terhadap sumbu y garis kedua

# mengecek kondisi persamaan garis
if c_07 != a_07 and g_07 != e_07:
  print(f"Gradien garis pertama (m1)= {m1_07}")
  print(f"Gradien garis kedua (m2)= {m2_07}")
  if m1_07 == m2_07:
    print("Kedua garis sejajar nih sehingga kemungkinannya ada dua: gak mempunyai solusi atau solusinya banyak")
    if c1_07 == c2_07:
      print("Kedua garis berhimpitan nih pastinya tak berhingga banyaknya solusi")
      if c1_07 and c2_07>0:
        print("Persamaan garis pertama adalah y1={}x+{} ".format(m1_07,c1_07))
        print("Persamaan garis kedua adalah y2={}x+{} ".format(m2_07,c2_07))
      else:
        print("Persamaan garis pertama adalah y1={}x{} ".format(m1_07,c1_07))
        print("Persamaan garis kedua adalah y2={}x{} ".format(m2_07,c2_07))
      x_07 = np.linspace(-5,5,100)
      x_07 = np.arange(-5,20,0.1)
      y_07 = np.arange(-20,20,0.1)
      plt.plot(x_07,m1_07*(x_07-a_07)+b_07,color='magenta',linestyle='--')
      plt.plot(x_07,m2_07*(x_07-e_07)+f_07,color='cyan')
      # memberi label untuk x dan y
      xlabel_07 = plt.xlabel('Sumbu X')
      ylabel_07 = plt.ylabel('Sumbu Y')
      # menambahkan garis kotak ke dalam plot
      grid_07 = plt.grid()
      plt.show()
    else:
      print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
      if c1_07 and c2_07>0:
        print("Persamaan garis pertama adalah y1={}x+{} ".format(m1_07,c1_07))
        print("Persamaan garis kedua adalah y2={}x+{} ".format(m2_07,c2_07))
      else:
        print("Persamaan garis pertama adalah y1={}x{} ".format(m1_07,c1_07))
        print("Persamaan garis kedua adalah y2={}x{} ".format(m2_07,c2_07))
      x_07 = np.linspace(-5,5,100)
      x_07 = np.arange(-5,20,0.1)
      y_07 = np.arange(-20,20,0.1)
      plt.plot(x_07,m1_07*(x_07-a_07)+b_07,color='pink')
      plt.plot(x_07,m2_07*(x_07-e_07)+f_07,color='red')
      # memberi label untuk x dan y
      xlabel_07 = plt.xlabel('Sumbu X')
      ylabel_07 = plt.ylabel('Sumbu Y')
      # menambahkan garis kotak ke dalam plot
      grid_07 = plt.grid()
      plt.show()

  if m1_07 * m2_07 == -1:
    print("Akan terdapat satu titik potong dan garis berpotongan tegak lurus")
    if c1_07 == c2_07:
      print("Titik potong terhadap sumbu y nya sama")
      if c1_07 and c2_07>0:
        print("Persamaan garis pertama adalah y1={}x+{} ".format(m1_07,c1_07))
        print("Persamaan garis kedua adalah y2={}x+{} ".format(m2_07,c2_07))
      else:
        print("Persamaan garis pertama adalah y1={}x{} ".format(m1_07,c1_07))
        print("Persamaan garis kedua adalah y2={}x{} ".format(m2_07,c2_07))
      # mencari titik potong
      tp_x_07 = ((m1_07 * a_07) - b_07 - (m2_07 * e_07) + f_07)/(m1_07 - m2_07)
      tp_y_07 = m2_07*(tp_x_07-e_07)+f_07
      print("Titik potong kedua garisnya adalah ({},{})".format(tp_x_07,tp_y_07))

      x_07 = np.linspace(tp_x_07-5,tp_x_07+5,100)
      x_07 = np.arange(-5,10,0.1)
      y_07 = np.arange(-5,10,0.1)
      plt.plot(x_07,m1_07*(x_07-a_07)+b_07,color='green',label='y={}x+{}'.format(m1_07,c1_07))
      plt.plot(x_07,m2_07*(x_07-e_07)+f_07,color='orange',label='y={}x+{}'.format(m2_07,c2_07))
      plt.plot(tp_x_07,tp_y_07,marker='.',ms='10',mfc='black',mec='black')
      plt.axvline(x=0,color='grey')
      plt.axhline(y=0,color='grey')
      # memberi label untuk x dan y
      xlabel_07 = plt.xlabel('Sumbu X')
      ylabel_07 = plt.ylabel('Sumbu Y')
      # menambahkan garis kotak ke dalam plot
      grid_07 = plt.grid()
      plt.show()
    else:
      print("Tidak berpotongan di sumbu x maupun sumbu y, akan dicari titik potongnya")
      if c1_07 and c2_07>0:
        print("Persamaan garis pertama adalah y1={}x+{} ".format(m1_07,c1_07))
        print("Persamaan garis kedua adalah y2={}x+{} ".format(m2_07,c2_07))
      else:
        print("Persamaan garis pertama adalah y1={}x{} ".format(m1_07,c1_07))
        print("Persamaan garis kedua adalah y2={}x{} ".format(m2_07,c2_07))
      # mencari titik potong
      tp_x_07 = ((m1_07 * a_07) - b_07 - (m2_07 * e_07) + f_07)/(m1_07 - m2_07)
      tp_y_07 = m2_07*(tp_x_07-e_07)+f_07
      print("Titik potong kedua garisnya adalah ({},{})".format(tp_x_07,tp_y_07))

      x_07 = np.linspace(tp_x_07-5,tp_x_07+5,100)
      x_07 = np.arange(-5,10,0.1)
      y_07 = np.arange(-5,10,0.1)
      plt.plot(x_07,m1_07*(x_07-a_07)+b_07,color='green',label='y={}x+{}'.format(m1_07,c1_07))
      plt.plot(x_07,m2_07*(x_07-e_07)+f_07,color='orange',label='y={}x+{}'.format(m2_07,c2_07))
      plt.plot(tp_x_07,tp_y_07,marker='.',ms='10',mfc='black',mec='black')
      plt.axvline(x=0,color='grey')
      plt.axhline(y=0,color='grey')
      # memberi label untuk x dan y
      xlabel_07 = plt.xlabel('Sumbu X')
      ylabel_07 = plt.ylabel('Sumbu Y')
      # menambahkan garis kotak ke dalam plot
      grid_07 = plt.grid()
      plt.show()
else:
  print("Tidak bisa dibagi dengan nol, coba masukkan ulang titik")
```
